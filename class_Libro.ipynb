```python
class Libro: #crear clase es como los planos
  def __init__(self, titulo, autor, isbn, precio):  #es el constructor
      self.titulo = titulo      #atributo del objeto
      self.autor = autor
      self.isbn = isbn
      self.precio = precio


mi_libro1 = Libro("Python", "Oscar Ruíz", "123-456-789", 100000)
mi_libro2 = Libro("HTML", "Lorena Barrios", "100-155-156", 200000)
mi_libro3 = Libro("JavaScrip", "María Hernandez", "101-102-103", 150000)
mi_libro4 = Libro("Programación", "Jesus Rey", "197-150-102", 170000)
mi_libro5 = Libro("Logica", "Magdalena Rios", "101-102-103", 150000)

#método

print("\nLos nombres de los libros son: ")
print(f"El nombre del primer libro es: {mi_libro1.titulo}")
print(f"El nombre del segundo libro es: {mi_libro2.titulo}")
print(f"El nombre del tercero libro es: {mi_libro3.titulo}")
print(f"El nombre del cuarto libro es: {mi_libro4.titulo}")
print(f"El nombre del quinto libro es: {mi_libro5.titulo}")

print("\nLos Autores son: ")
print(f"El autor del primer libro es: {mi_libro1.autor}")
print(f"El autor del segundo libro es: {mi_libro2.autor}")
print(f"El autor del tercero libro es: {mi_libro3.autor}")
print(f"El autor del cuarto libro es: {mi_libro4.autor}")
print(f"El autor del quinto libro es: {mi_libro5.autor}")

print("\nLos isbn de los libros son: ")
print(f"El isbn del primer libro es: {mi_libro1.isbn}")
print(f"El isbn del segundo libro es: {mi_libro2.isbn}")
print(f"El isbn del tercero libro es: {mi_libro3.isbn}")
print(f"El isbn del cuarto libro es: {mi_libro4.isbn}")
print(f"El isbn del quinto libro es: {mi_libro5.isbn}")

print("\nLos precios de cada libro son: ")
print(f"El precio del primer libro es: {mi_libro1.precio}")
print(f"El precio del segundo libro es: {mi_libro2.precio}")
print(f"El precio del tercero libro es: {mi_libro3.precio}")
print(f"El precio del cuarto libro es: {mi_libro4.precio}")
print(f"El precio del quinto libro es: {mi_libro5.precio}")


print("\nLos datos de todos los libros son: ")
print(f"\nTitulo: {mi_libro1.titulo} \nAutor: {mi_libro1.autor} \nIsbn: {mi_libro1.isbn}  \nPrecio: {mi_libro1.precio}")
print(f"\nTitulo: {mi_libro2.titulo} \nAutor: {mi_libro2.autor} \nIsbn: {mi_libro2.isbn}  \nPrecio: {mi_libro2.precio}")
print(f"\nTitulo: {mi_libro3.titulo} \nAutor: {mi_libro3.autor} \nIsbn: {mi_libro3.isbn}  \nPrecio: {mi_libro3.precio}")
print(f"\nTitulo: {mi_libro4.titulo} \nAutor: {mi_libro4.autor} \nIsbn: {mi_libro4.isbn}  \nPrecio: {mi_libro4.precio}")
print(f"\nTitulo: {mi_libro5.titulo} \nAutor: {mi_libro5.autor} \nIsbn: {mi_libro5.isbn}  \nPrecio: {mi_libro5.precio}")


print("\nLos libros que tenemos actualmente son: ")
print(f"\nTitulo del Libro 1: {mi_libro1.titulo} \nTitulo del Libro 2: {mi_libro2.titulo} \nTitulo del Libro 3: {mi_libro3.titulo} \nTitulo del Libro 4: {mi_libro4.titulo} \nTitulo del Libro 5: {mi_libro5.titulo}")

print("\nAutor de cada libro: ")
print(f"\nAutor del Libro 1: {mi_libro1.autor} \nAutor del Libro 2: {mi_libro2.autor} \nAutor del Libro 3: {mi_libro3.autor} \nAutor del Libro 4: {mi_libro4.autor} \nAutor del Libro 5: {mi_libro5.autor}")

print("\nIsbn de los libros: ")
print(f"\nIsbn del Libro 1: {mi_libro1.isbn} \nIsbn del Libro 2: {mi_libro2.isbn} \nIsbn del Libro 3: {mi_libro3.isbn} \nIsbn del Libro 4: {mi_libro4.isbn} \nIsbn del Libro 5: {mi_libro5.isbn}")

print("\nEl Precio de cada libro es: ")
print(f"\nPrecio del Libro 1: {mi_libro1.precio} \nPrecio del Libro 2: {mi_libro2.precio} \nPrecio del Libro 3: {mi_libro3.precio} \nPrecio del Libro 4: {mi_libro4.precio} \nPrecio del Libro 5: {mi_libro5.precio}")




```
