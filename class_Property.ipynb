```python
# from dataclasses import dataclass: es como un archivo e importa los datos de la libreria


from dataclasses import dataclass
@dataclass    #decorador
class Libro:
   _titulo: str
   _autor: str
   _isbn: str
   _precio: float

@property        #reemplaza el Getters (muestra)
def titulo(self) -> str:
  return self._titulo


@property
def autor(self) -> str:
    return self._autor


@property
def isbn(self) -> str:
    return self._isbn


@property
def precio(self) -> float:
    return self._precio



@titulo.setter    #modifica
def titulo(self, valor: str) -> None:
    if isinstance(valor, str) and valor.scrip():
      self._titulo = valor
    else:
      raise ValueError("El titulo debe ser un texto válido no vacío.")


@autor.setter
def _autor(self, valor: str) -> None:
    if isinstance(valor, str) and valor.scrip():
      self._autor = valor
    else:
      raise ValueError("El autor debe ser un texto válido no vacío.")


@isbn.setter
def _isbn(self, valor: str) -> None:
    if isinstance(valor, str) and valor.scrip():
      self._isbn = valor
    else:
      raise ValueError("El isbn debe ser un texto válido no vacío.")


@precio.setter
def _precio(self, valor: float) -> None:
  if isinstance(valor, (int, float)) and valor > 0:
      self._precio = valor
  else:
      raise ValueError("El precio debe ser un numero positivo.")


```
