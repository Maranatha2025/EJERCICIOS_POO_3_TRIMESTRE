```python
#Aquí empiezo escribiendo mi código python orientada a objetos (POO)
#En este ejercicio de class nos piden mostrar la información de cada estudiante de diferentes maneras


class Estudiantes: #clase Estudiantes es el molde como los planos
    def __init__(self, nombre, edad, peso ,estatura ,documento, genero): #Es el constructor de la clase y en el encontramos los objetos
       self.nombre= nombre   #atributo 1, self representa el objeto actual, es decir, la instancia que se está usando en ese momento.
       self.edad = edad
       self.peso = peso
       self.estatura =  estatura
       self.documento = documento
       self.genero = genero

#Creo los objetos de la clase padre
estudiante1 = Estudiantes("CARLOS","20","70 kg","1.76","1234567891","F") #Ingreso los datos directamente
estudiante2 = Estudiantes("KARLA","20","60 kg","1.79","1234567892","M")
estudiante3 = Estudiantes("ARNOL","18","76 kg","1.77","1234567893","M")
estudiante4 = Estudiantes("LORENA","25","56 kg","1.59","1234567894","F")
estudiante5 = Estudiantes("LUIS","21","80 kg","1.65","1234567895","M")
estudiante6 = Estudiantes("ALBERTO","19","78 kg","2.00","1234567897","F")

#Muestro la información de los estudiantes el nombre y el peso
print("\nAQUI SE MUESTRA EL NOMBRE Y EL PESO DE LOS ESTUDIANTES:\n") #print: nos permite mostrar los datos, \n nos permite un salto de linea
print(f"\nNOMBRE:{estudiante1.nombre} \nPESO:{estudiante1.peso}") #La f sirve para poner variables {muestra el objeto} dentro de un texto de forma fácil.
print(f"\nNOMBRE:{estudiante2.nombre} \nPESO:{estudiante2.peso}")
print(f"\nNOMBRE:{estudiante3.nombre} \nPESO:{estudiante3.peso}")
print(f"\nNOMBRE:{estudiante4.nombre} \nPESO:{estudiante4.peso}")
print(f"\nNOMBRE:{estudiante5.nombre} \nPESO:{estudiante5.peso}")
print(f"\nNOMBRE:{estudiante6.nombre} \nPESO:{estudiante6.peso}")

print("\nAQUI SE MUESTRA EL DOCUMENTO, LA ESTATURA Y LA EDAD DE LOS ESTUDIANTES:\n")
print(f"\nDOCUMENTO:{estudiante1.documento} \nESTATURA:{estudiante1.estatura} \nEDAD:{estudiante1.edad}")
print(f"\nDOCUMENTO:{estudiante2.documento} \nESTATURA:{estudiante2.estatura} \nEDAD:{estudiante2.edad}")
print(f"\nDOCUMENTO:{estudiante3.documento} \nESTATURA:{estudiante3.estatura} \nEDAD:{estudiante3.edad}")
print(f"\nDOCUMENTO:{estudiante4.documento} \nESTATURA:{estudiante4.estatura} \nEDAD:{estudiante4.edad}")
print(f"\nDOCUMENTO:{estudiante5.documento} \nESTATURA:{estudiante5.estatura} \nEDAD:{estudiante5.edad}")
print(f"\nDOCUMENTO:{estudiante6.documento} \nESTATURA:{estudiante6.estatura} \nEDAD:{estudiante6.edad}")

print("\nAQUI SE MUESTRA CADA NOMDRE DE LOS ESTUDIANTES:\n")
print(f"\nNOMBRE DEL PRIMER ESTUDIANTE:{estudiante1.nombre} \nNOMBRE DEL SEGUNDO ESTUDIANTE:{estudiante2.nombre} \nNOMBRE DEL TERCER ESTUDIANTE:{estudiante3.nombre} \nNOMBRE DEL CUARTO ESTUDIANTE:{estudiante4.nombre} \nNOMBRE DEL QUINTO ESTUDIANTE:{estudiante5.nombre} \nNOMBRE DEL SEXTO ESTUDIANTE:{estudiante6.nombre}")


print("\nAQUI SE MUESTRA EL NOMBRRE Y EL GENERO DE LOS ESTUDIANTES")
print(f"\nNOMBRE:{estudiante1.nombre} \nGENERO:{estudiante1.genero}")
print(f"\nNOMBRE:{estudiante2.nombre} \nGENERO:{estudiante2.genero}")
print(f"\nNOMBRE:{estudiante3.nombre} \nGENERO:{estudiante3.genero}")
print(f"\nNOMBRE:{estudiante4.nombre} \nGENERO:{estudiante4.genero}")
print(f"\nNOMBRE:{estudiante5.nombre} \nGENERO:{estudiante5.genero}")
print(f"\nNOMBRE:{estudiante6.nombre} \nGENERO:{estudiante6.genero}")
```
