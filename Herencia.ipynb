```python
#HERENCIA: permite crear una clase nueva (subclase) a partir de otra (superclase) que es la clase principal,
#la subclase reutiliza los atributos y metodos de la superclase

#la clase PADRE es una clase base que contiene atributos o comportamiento comunes
#la clase HIJA hereda los comportamientos puede agregar o modificar lo que necesite

# crear clase Padre
class Vehiculo:
  def mover(self):  #Metodo
    print("El vehiculo se esta moviendo")

# crear clases hijas
class Moto(Vehiculo):  #hereda el metodo mover de la clase Padre del Vehiculo
  def mover(self):     #Aqui se sobre escribe el metodo mover()
    print("El carro se esta moviendo")

class Avion(Vehiculo):
  def mover(self):
    print("El vehiculo se esta moviendo")

class Barco(Vehiculo):
  def mover(self):
    print("El vehiculo se esta moviendo")

class Submarino(Vehiculo):
  def mover(self):
    print("El vehiculo se esta moviendo")

def main(): #Aqui inicia el programa
  #creo mi primer OBJETO clase Padre
  print("\nAqui muestro el metodo de la clase padre\n")
  vehiculo1 = Vehiculo() #Creo un objeto de la clase padre
  vehiculo1.mover()   #Llamo al metodo mover del objeto de la clase

  #creo mi primer OBJETO clase hijo
  print("\nLa clase hijo repetira el metodo mover () de la clase padre\n")
  #Aqui nos muestra sel sobrescritura del metodo mover() de la clase moto
  moto1 = Moto()
  moto1.mover()  #Llamo al metodo mover del objeto de la clase

  #creo mi segundo OBJETO clase hijo
  print("\nLa clase hijo repetira el metodo mover () de la clase padre\n")
  avion1 = Avion()
  avion1.mover()

  #creo mi tercer OBJETO clase hijo
  print("\nLa clase hijo repetira el metodo mover () de la clase padre\n")
  barco1 = Barco()
  barco1.mover()

  #creo mi cuarto OBJETO clase hijo
  print("\nLa clase hijo repetira el metodo mover () de la clase padre\n")
  submarino1 = Submarino()
  submarino1.mover()

if __name__ == "__main__":
     main()




```

```python
class Padre: #Clase
  def __init__(self, mensaje, color_ojos, tipo_sangre): #Constructor
    self.mensaje = mensaje
    self.color_ojos = color_ojos
    self.tipo_sangre = tipo_sangre


class Hijo(Padre):
  def __init__(self, mensaje, color_ojos, tipo_sangre, nombre, genero, estatura):
    super().__init__(mensaje, color_ojos, tipo_sangre)  #Reutiliza el constructor de la clase padre
    self.nombre = nombre
    self.genero = genero
    self.estatura = estatura


def main():
    obj = Hijo("Hola desde la clase Padre", "Tengo los ojos Azules", "O-", "Me llamo  jesús","Masculino", "1.89") #instanciamos un objeto de la clase hijo
    print(f"{obj.mensaje}, {obj.color_ojos}, {obj.tipo_sangre}, {obj.nombre}, {obj.genero}, {obj.estatura}\n")
    print(obj.mensaje)
    print(obj.color_ojos)
    print(obj.tipo_sangre)
    print(obj.genero)
    print(obj.estatura)


if __name__ == "__main__":
      main()
```
