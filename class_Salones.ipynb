```python
class Salones:
  def __init__(self, codigo, nombre, responsable, puestos):
      self.codigo = codigo
      self.nombre = nombre
      self.responsable = responsable
      self.puestos = puestos

salones1 = Salones("1", "MATEMATICAS", "OSCAR", "50")
salones2 = Salones("2", "INGLES", "ALIZ", "40")
salones3 = Salones("3", "ALGEBRA", "LUIS", "30")
salones4 = Salones("4", "DANZAS", "LEO", "70")
salones5 = Salones("1", "RELIGION", "MARÍA", "20")

print(f"\nLa organización de cada uno de los Salones es: ")
print(f"\nCODIGO:{salones1.codigo} \nNOMBRE:{salones1.nombre} \nRESPONSABLE:{salones1.responsable} \nPUESTOS:{salones1.puestos}")
print(f"\nCODIGO:{salones2.codigo} \nNOMBRE:{salones2.nombre} \nRESPONSABLE:{salones2.responsable} \nPUESTOS:{salones2.puestos}")
print(f"\nCODIGO:{salones3.codigo} \nNOMBRE:{salones3.nombre} \nRESPONSABLE:{salones3.responsable} \nPUESTOS:{salones3.puestos}")
print(f"\nCODIGO:{salones4.codigo} \nNOMBRE:{salones4.nombre} \nRESPONSABLE:{salones4.responsable} \nPUESTOS:{salones4.puestos}")
print(f"\nCODIGO:{salones5.codigo} \nNOMBRE:{salones5.nombre} \nRESPONSABLE:{salones5.responsable} \nPUESTOS:{salones5.puestos}")
```
