```python
class Automovil:   #clase
    def __init__(self, marca, modelo, color):
      self.marca = marca
      self.modelo = modelo
      self.color = color

    def mostrar_info(self):     #método
      print(f"\nLa marca del vehiculo es: {self.marca}")
      print(f"El modelo del vehiculo es: {self.modelo}")
      print(f"El color del vehiculo es: {self.color}")

    def main():
      marca = input("Ingrese la marca del vehiculo: ")
      modelo = input("Ingrese el modelo del vehiculo: ")
      color = input("Ingrese el color del vehiculo: ")
      automovil1 = Automovil(marca, modelo, color)
      automovil2 = Automovil("Pulsar", 2026, "Azul") #Ingreso los datos directamente
      automovil1.mostrar_info()
      automovil2.mostrar_info()
    if __name__ == "__main__":
        main()
```

```python
class Perro:   #clase
    def __init__(self, nombre, raza, edad, color, peso):  #Constructor
      self.nombre = nombre      #atributo 1
      self.raza = raza          #atributo 2
      self.edad = edad        #atributo 3
      self.color = color          #atributo 4
      self.peso = peso          #atributo 5

    def ladrar(self):     #método
      print(f"{self.nombre} dice: Guau guau!")


    def mostrar_info(self):
      print(f"\nInformación del perro:")
      print(f"El nombre del perro es: {self.nombre}")
      print(f"La raza del perro es: {self.raza}")
      print(f"La edad del perro es: {self.edad} años")
      print(f"El color del perro es: {self.color}")
      print(f"El peso del perro es: {self.peso} kg")


def main():
    nombre = input("Ingrese el nombre del perro: ")
    raza = input("Ingrese la raza del perro: ")
    edad = input("Ingrese la edad del perro: ")
    color = input("Ingrese el color del perro: ")
    peso = input("Ingrese el peso del perro: ")

    perro1 = Perro(nombre, raza, edad, color, peso)
    perro2 = Perro("Luna", "Lobo", 15, "Cafe", 20.6)
    perro1.mostrar_info()
    perro1.ladrar()
    perro2.mostrar_info()
    perro2.ladrar()

if __name__ == "__main__":
        main()
```

```python
class Cliente:   #clase
    def __init__(self, nombre, apellido, edad, sexo, peso, direccion, telefono, estado_civil, color_ojos, color_cabello):  #Constructor
      self.nombre = nombre               #atributo 1
      self.apellido = apellido           #atributo 2
      self.edad = edad                   #atributo 3
      self.sexo = sexo                   #atributo 4
      self.peso = peso                   #atributo 5
      self.direccion= direccion          #atributo 6
      self.telefono = telefono           #atributo 7
      self.estado_civil = estado_civil   #atributo 8
      self.color_ojos = color_ojos       #atributo 9
      self.color_cabello = color_cabello #atributo 10


    def Saludo(self):     #método
      print(f"{self.nombre} dice: ¡Hola!")


    def mostrar_info(self):
      print(f"\nInformación del cliente:")
      print(f"El nombre del cliente es: {self.nombre}")
      print(f"El apellido del cliente es: {self.apellido}")
      print(f"La edad del cliente es: {self.edad} años")
      print(f"El sexo del cliente es: {self.sexo}")
      print(f"El peso del perro es: {self.peso} kg")
      print(f"La direccion del cliente es: {self.direccion}")
      print(f"El telefono del cliente es: {self.telefono}")
      print(f"El estado civil del cliente es: {self.estado_civil}")
      print(f"El color de ojos del cliente es: {self.color_ojos}")
      print(f"El color del cabello del cliente es: {self.color_cabello}")


def main():
    nombre = input("Ingrese el nombre del cliente: ")
    apellido = input("Ingrese el apellido del cliente: ")
    edad = input("Ingrese la edad del cliente: ")
    sexo = input("Ingrese el sexo del cliente: ")
    peso = input("Ingrese el peso del cliente: ")
    direccion = input("Ingrese la direccion del cliente: ")
    telefono = input("Ingrese el telefono del cliente: ")
    estado_civil = input("Ingrese el estado civil del cliente: ")
    color_ojos = input("Ingrese el color de ojos del cliente: ")
    color_cabello = input("Ingrese el color del cabello del cliente: ")

    cliente1 = Cliente(nombre, apellido, edad, sexo, peso, direccion, telefono, estado_civil, color_ojos, color_cabello)
    cliente2 = Cliente("Roxana", "Ruiz", 26, "f", 56.8, "Calle 123", 123456789, "soltera", "marron", "negro")
    cliente1.mostrar_info()
    cliente2.Saludo()
    cliente1.mostrar_info()
    cliente2.Saludo()

if __name__ == "__main__":
        main()
```
