```python
class Empleado:  #clase Padre (principal)
  def __init__(self, nombre, salario):  #constructor de la plase padre
   self.nombre = nombre       #Atributos1
   self.salario = salario     #Atributos2

class Gerente(Empleado):
  def __init__(self, nombre, salario, departamento):
    super().__init__(nombre, salario)
    self.departamento = departamento

  def mostrar_info(self):
    print(f"Nombre:{self.nombre}, Salario:${self.salario:,.0f}")
    print(f"Departamento:{self.departamento}, Salario:${self.salario:,.0f}")

def main():
  gerente1 = Gerente("Roxana Ruiz", 1759000 ,"Innovación")
  gerente1.mostrar_info()


if __name__ == "__main__":
  main()
```
