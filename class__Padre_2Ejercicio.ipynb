```python
class Padre:
  def __init__(self, mensaje):
    self.mensaje = mensaje

class Hijo(Padre):
  def __init__(self, mensaje, nombre):
    super().__init__(mensaje)
    self.nombre = nombre

def main():
    obj = Hijo("Hola desde la clase padre","Roxana")
    print(f"{obj.mensaje},{obj.nombre} \n")
    print(obj.mensaje)
    print(obj.nombre)

if __name__ == "__main__":
  main()



```
