```python
#class getters y setters def __init__()_: Ejercicio de practica crear si donde se modifica y se muestra información de prendas.
#pide los valores de cambio por print("")

#Público	atributo	self.titulo	Puede accederse desde cualquier parte.
#Protegido	_atributo	self._precio	Se usa por convención para acceso interno.
#Privado	__atributo	self.__precio	Se “oculta” del exterior mediante name mangling.


class Prenda:
    def __init__(self, nombre, talla, color, precio):
        self._nombre = nombre
        self._talla = talla
        self._color = color
        self._precio = precio

#getters Obtener un valor

    def get_nombre(self):
        return self._nombre     #Devuelve el valor

    def get_talla(self):
        return self._talla

    def get_color(self):
        return self._color

    def get_precio(self):
        return self._precio



#setters controla y valida antes de modificar el valor

    def set_nombre(self, nombre):
        self._nombre = nombre

    def set_talla(self, talla):
        self._talla = talla

    def set_color(self, color):
        self._color = color

    def set_precio(self, precio):
        if precio > 0:
            self._precio = precio
        else:
            print("\nEl precio debe ser mayor a 0")

#Mostrar información de la prenda

    def mostrar_info(self):
        print("\nInformación de la prenda")
        print(f"Nombre: {self._nombre}")
        print(f"Talla: {self._talla}")
        print(f"Color: {self._color}")
        print(f"Precio: ${self._precio}")


#Crear objeto

prenda1 = Prenda("Blusa", "S", "Blanco", 20000)
prenda1.mostrar_info()


# Modificar datos usando input

print("\nModificar información de la prenda")

nuevo_nombre = input("Ingrese nuevo nombre: ")
nueva_talla = input("Ingrese nueva talla: ")
nuevo_color = input("Ingrese nuevo color: ")
nuevo_precio = float(input("Ingrese nuevo precio: "))

# Usar setters

prenda1.set_nombre(nuevo_nombre)
prenda1.set_talla(nueva_talla)
prenda1.set_color(nuevo_color)
prenda1.set_precio(nuevo_precio)


# Mostrar datos modificados
prenda1.mostrar_info()
```
