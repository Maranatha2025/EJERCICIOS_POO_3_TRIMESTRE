```python
#pilares de POO herencia abstracción encapsulamiento polimorfismo
#en POO se llama atributos a las (variables) usando {}.

#__init__ es el constructor.
#Se ejecuta automáticamente al crear el objeto.
#self representa el objeto.
#titulo y precio son datos de entrada.

class Libro: #class crea una clase. Libro es el nombre de la clase. Es una plantilla para crear objetos. Libro es el molde, libro1 es el objeto.
  def __init__(self, titulo, precio):  #Constructor def __init__(self, titulo, precio):
    self.__titulo = titulo       #El doble __ representa atributo privado, guarda el tt en el atributo privado
    self.__precio = precio       # titulo y precio son datos de entrada

#Decoradores @property permite acceder a un método como si fuera un atributo (getter moderno)
#"""Getter profesional (permite acceder como si fuera un atributo).""", el getter: obtener (leer) el valor de un atributo

  @property     # Decorador con este comando creo el getter moderno (get) la información de los atributos de cada clase
  def titulo(self):
    return self.__titulo


  @property  # Decorador con este comando creo el getter moderno (get) la información de los atributos de cada clase
  def precio(self):
   return self.__precio   #Devuelve el precio.


  @titulo.setter    # Decorador
  def titulo(self, nuevo_titulo):
    if isinstance(nuevo_titulo, str):  #isinstance() = sirve para verificar si una variable es de un tipo específico.
      self.__titulo = nuevo_titulo
    else:
      print("Debe ingresar un texto valido en el titulo")

  @precio.setter  # modificar
  def precio(self, nuevo_precio):
      if isinstance(nuevo_precio, (int, float)) and nuevo_precio > 0:  #Es el objeto real creado usando ese molde
         self.__precio = nuevo_precio
      else:
         print("Debe ingresar un numero valido en el precio")

  def mostrar_informacion(self):   #Método para mostrar información.
    print(f"El titulo del libro es: {self.__titulo}") #Muestra el título.
    print(f"El precio del libro es: {self.__precio}") #Muestra el precio.
    #otra forma de mostrar la información junta llamando con f
    print(f"El titulo del libro es: {self.__titulo}, el precio del libro es {self.__precio}")  #\n: sirve para dar un salto de linea


def main(): #main es una funcion de python organiza el programa. () van los parametros: empieza el codigo
  print("=======SISTEMA DE LIBROS (POO + Encapsulamiento)=======\n")
  libro1 = Libro("El PRINCIPITO", 60000)   #creación del objeto
  libro2 = Libro("El QUIJOTE DE LA MANCHA", 20000)

#Mostrar información inicial
  print("DATOS DE LOS LIBROS:\n")

  #Datos del libro 1
  print("El nombre del libro 1 es:", libro1.titulo)
  print("El precio del libro 1 es:", libro1.precio)
  print("El nombre del libro 1 y su precio es:", libro1.titulo, "Precio:", libro1.precio)

  print("\nTODOS LOS DATOS DEL LIBRO 1\n")
  libro1.mostrar_informacion()

  print("\nActualizaremos el titulo del libro 1\n")
  libro1.titulo = input("Ingrese el nuevo titulo:")
  print("Titulo actualizado ahora el titulo del libro 1 es:",libro1.titulo)

  print("\nActualizarenos el precio del libro 1\n")
  libro1.precio = float(input("Ingrese el nuevo precio del libro 1:"))
  print("\nPrecio actualizado ahora el precio del libro 1 es:",libro1.precio)



  #Datos del libro 2
  print("\nEl nombre del libro 2 es:", libro2.titulo)
  print("El precio del libro 2 es:", libro2.precio)
  print("El nombre del libro 2 y su precio es:", libro2.titulo, "Precio:", libro2.precio)

  print("\nTODOS LOS DATOS DEL LIBRO 2\n")
  libro2.mostrar_informacion()

  print("\nActualizaremos el titulo del libro 2\n")
  libro2.titulo = input("Ingrese el nuevo titulo:")
  print("Titulo actualizado ahora el titulo del libro 2 es:",libro2.titulo)

  print("\nActualizarenos el precio del libro 2\n")
  libro2.precio = float(input("Ingrese el nuevo precio del libro 2:"))
  print("\nPrecio actualizado ahora el precio del libro 2 es:",libro2.precio)



if __name__ == "__main__": #METODO PRINCIPAL QUE LLAMA AL METODO DE ENTRADA
  main()
```
