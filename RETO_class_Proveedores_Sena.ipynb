```python
#Aplicar la técnica tradicional y Decoradores para crear las dos soluciones del sistema

#Ejercicio de practica crear un si para proveedores, de la empresa Sena

#Registrar 3 proveedores/ la clase debe tener 5 atríbutos/ deben tener TYPADO FUERTE, con decoradores/ Actualizar de la información de
#3 atributos seleccionados voluntariamente de los 5 atríbutos/
#utilizar los métodos para mostrar las intancias de cada atríbuto


from dataclasses import dataclass

@dataclass
class Proveedores_Sena:
    _nombre: str
    _nit: str
    _tel: str | None = None
    _dir: str | None = None
    _correo: str | None = None

    # ---------- GETTERS ----------
    @property
    def nombre(self) -> str:
        return self._nombre

    @property
    def nit(self) -> str:
        return self._nit

    @property
    def tel(self) -> str | None:
        return self._tel

    @property
    def dir(self) -> str | None:
        return self._dir

    @property
    def correo(self) -> str | None:
        return self._correo

    # ---------- SETTERS ----------
    @nombre.setter
    def nombre(self, valor: str) -> None:
        if isinstance(valor, str) and valor.strip():
            self._nombre = valor.strip()
        else:
            raise ValueError("El nombre debe ser un texto válido y no vacío.")

    @tel.setter
    def tel(self, valor: str) -> None:
        if isinstance(valor, str) and valor.strip():
            self._tel = valor.strip()
        else:
            raise ValueError("El teléfono debe ser un texto válido y no vacío.")

    @dir.setter
    def dir(self, valor: str) -> None:
        if isinstance(valor, str) and valor.strip():
            self._dir = valor.strip()
        else:
            raise ValueError("La dirección debe ser un texto válido y no vacío.")

    # ---------- REPRESENTACIÓN ----------
    def __repr__(self) -> str:
        return (
            f"Proveedor_Sena("
            f"nombre='{self._nombre}', "
            f"nit='{self._nit}', "
            f"tel='{self._tel}', "
            f"dir='{self._dir}', "
            f"correo='{self._correo}')"
        )


# ---------- FUNCIÓN MAIN ----------
def main() -> None:
    # Registrar 3 proveedores
    p1 = Proveedores_Sena(
        "Proveedor Uno",
        "900001",
        "3001111111",
        "Bogotá",
        "uno@sena.edu.co"
    )

    p2 = Proveedores_Sena(
        "Proveedor Dos",
        "900002",
        None,
        None,
        "dos@sena.edu.co"
    )

    p3 = Proveedores_Sena(
        "Proveedor Tres",
        "900003"
    )

    print("PROVEEDORES REGISTRADOS")
    print(p1)
    print(p2)
    print(p3)

    # Actualizar 3 atributos
    p1.nombre = "Proveedor Uno Actualizado"
    p1.tel = "3119999999"
    p1.dir = "Medellín"

    print("\nDESPUÉS DE ACTUALIZAR")
    print(p1)


# ---------- EJECUCIÓN ----------
if __name__ == "__main__":
    main()












```
